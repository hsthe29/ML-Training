# Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(11)
```

```python
X = 100 * np.random.rand(1000, 1)
y = 1.2785 * X + 10 + 5 * np.random.randn(1000, 1)

one = np.ones_like(X)
X_bar = np.concatenate((one, X), axis = 1)

Ap = X_bar.T @ X_bar
b = X_bar.T @ y

w = np.linalg.pinv(Ap) @ b

x0 = np.linspace(0, 100, 2)
y0 = w[0, 0] + w[1, 0] * x0

print(w.T)

plt.plot(X, y, 'b.')
plt.plot(x0, y0, 'r', linewidth = 4)

plt.show()
```

# Using Scikit Learn

```python
from sklearn import linear_model

regr = linear_model.LinearRegression(fit_intercept=False)
regr.fit(X_bar, y)

print(regr.coef_)
```
