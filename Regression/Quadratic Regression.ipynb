# Function

Original function: $f(x) = x^2 + 2x + 3$

$f(x) = ax^2 + bx + c$

# Manual

```python
# import modules

import numpy as np
import matplotlib.pyplot as plt
```

```python
def org_func(x):
  return x ** 2 + 2 * x + 3
```

```python
X = np.array([np.arange(-20, 20, 0.1)]).T
X_sq = np.asarray(list(map(lambda x: x * x, X)))
y = org_func(X) + np.array([np.random.randn(X.shape[0])]).T

one = np.ones_like(X) # biases
X_bar = np.concatenate((one, X, X_sq), axis = 1)
```

```python
Ap = X_bar.T @ X_bar
b = X_bar.T @ y

w = np.linalg.pinv(Ap) @ b
print(w)
```

```python
a = w[2][0]
b = w[1][0]
c = w[0][0]

x_o = np.linspace(-4, 5, 100)
y_o = org_func(x_o)

pred_func = a * x_o ** 2 + b * x_o + c

plt.plot(X[160:250], y[160:250], 'bo', label = 'Original data', markersize = 4)
plt.plot(x_o, pred_func, 'r', label = 'Fitted graph')

plt.title('a part of chart')
plt.legend()
plt.show()
```

# Using Scikit learn

```python
from sklearn import datasets, linear_model

regr = linear_model.LinearRegression(fit_intercept=False)
regr.fit(X_bar, y)

print( 'Solution found by scikit-learn  : ', regr.coef_ )
print( 'Solution found by manual: ', w.T)
```
